Implement various types of recurrent neural network architecture in tensorflow ,including one-to-one, many- to-one, many-to-many, one-to-many . for each architecture build a model step by step by defining the input layer, recurrent layer and output layer. Compile and generate the model summary


s1- take all the req library
s2- take the dummy data from Kaggle etc
s3- normalise the data
s4- add layering, input, recurrence and output
s5- build model
s6- compile model
s7- show summary

```python
# Step 1: Import Libraries
import tensorflow as tf
import numpy as np

from tensorflow.keras.layers import Input, LSTM, Dense, TimeDistributed, RepeatVector
from tensorflow.keras.models import Model

# Step 2: Create Dummy Data
X = np.random.rand(100, 20, 10)   # (samples, timesteps, features)
y = np.random.rand(100, 1)

# Step 3: Normalize Data
X = (X - X.min()) / (X.max() - X.min())
y = (y - y.min()) / (y.max() - y.min())

# ================================
# 🔹 MANY-TO-ONE MODEL
# ================================

print("\n===== MANY-TO-ONE MODEL =====")

inputs = Input(shape=(20, 10))     # Input layer
x = LSTM(32)(inputs)               # Recurrent layer
outputs = Dense(1)(x)              # Output layer

model = Model(inputs, outputs)

model.compile(optimizer='adam', loss='mse')

model.summary()

# ================================
# 🔹 ONE-TO-MANY MODEL
# ================================

print("\n===== ONE-TO-MANY MODEL =====")

inputs = Input(shape=(10,))
x = Dense(32)(inputs)
x = RepeatVector(5)(x)
x = LSTM(32, return_sequences=True)(x)
outputs = TimeDistributed(Dense(1))(x)

model = Model(inputs, outputs)

model.compile(optimizer='adam', loss='mse')

model.summary()

# ================================
# 🔹 MANY-TO-MANY MODEL
# ================================

print("\n===== MANY-TO-MANY MODEL =====")

inputs = Input(shape=(20, 10))
x = LSTM(32, return_sequences=True)(inputs)
outputs = TimeDistributed(Dense(1))(x)

model = Model(inputs, outputs)

model.compile(optimizer='adam', loss='mse')

model.summary()

# ================================
# 🔹 ONE-TO-ONE MODEL
# ================================

print("\n===== ONE-TO-ONE MODEL =====")

inputs = Input(shape=(10,))
x = Dense(32, activation='relu')(inputs)
outputs = Dense(1)(x)

model = Model(inputs, outputs)

model.compile(optimizer='adam', loss='mse')

model.summary()
```
