```python
# =========================
# STEP 0: Install TensorFlow (only needed in some environments)
# =========================
!pip install tensorflow

# =========================
# STEP 1: Import Libraries
# =========================
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# =========================
# STEP 2: Create Dataset
# =========================
X = np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5],
    [4,5,6],
    [5,6,7],
    [6,7,8],
    [7,8,9],
    [8,9,10]
])

y = np.array([4,5,6,7,8,9,10,11])

# =========================
# STEP 3: Reshape (FIXED)
# =========================
X = X.reshape((X.shape[0], X.shape[1], 1))
print("X shape:", X.shape)

# =========================
# STEP 4: Model (YOUR CODE FIXED)
# =========================
model = Sequential()

model.add(LSTM(50, activation='relu', input_shape=(3,1)))
model.add(Dense(1))

# =========================
# STEP 5: Compile
# =========================
model.compile(optimizer='adam', loss='mse')

# =========================
# STEP 6: Summary
# =========================
model.summary()

# =========================
# STEP 7: Train
# =========================
model.fit(X, y, epochs=200, verbose=1)

```

```python
# =========================
# Step 1: Import Libraries
# =========================
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# =========================
# Step 2: Text Data
# =========================
text = """
deep learning models are powerful
deep learning models are complex
machine learning is fun
machine learning is powerful
deep models learn patterns
learning models need data
"""

# =========================
# Step 3: Tokenization
# =========================
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

word_index = tokenizer.word_index
vocab_size = len(word_index) + 1

# =========================
# Step 4: Create Sequences
# =========================
sequences = []

for line in text.strip().split("\n"):
    token_list = tokenizer.texts_to_sequences([line])[0]

    for i in range(3, len(token_list)):
        n_gram = token_list[i-3:i+1]   # 3 input + 1 output
        sequences.append(n_gram)

sequences = np.array(sequences)

# =========================
# Step 5: Split X and y
# =========================
X = sequences[:, :-1]
y = sequences[:, -1]

# One-hot encode y
y = to_categorical(y, num_classes=vocab_size)

# =========================
# Step 6: Build Model
# =========================
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=3))
model.add(LSTM(50))
model.add(Dense(vocab_size, activation='softmax'))

# =========================
# Step 7: Compile
# =========================
model.compile(loss='categorical_crossentropy', optimizer='adam')

model.summary()

# =========================
# Step 8: Train
# =========================
model.fit(X, y, epochs=200, verbose=1)

# =========================
# Step 9: Prediction Function
# =========================
def predict_next_word(seed_text):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = token_list[-3:]

    token_list = np.array(token_list).reshape(1, 3)

    predicted = model.predict(token_list, verbose=0)
    predicted_index = np.argmax(predicted)

    for word, index in word_index.items():
        if index == predicted_index:
            return word

# =========================
# Step 10: Test
# =========================
# test_inp = "deep learning models"
test_inp = "learning models are"

print("Input:", test_inp)
print("Next word:", predict_next_word(test_inp))
```
