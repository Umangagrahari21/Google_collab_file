```python
import nltk
nltk.download('brown')
```

```python
from nltk.corpus import brown
categories=brown.categories()
print(categories)
```

```python
sample_categories=['news','fiction','romance']
for sat in sample_categories:
  words=brown.words(categories=sat)
  print(f"\nCategory:{sat}")
  print(words[:20])
```

```python
for sat in brown.categories():
  print(sat, ":", len(brown.words(categories=sat)))
```

```python
news_words=brown.words(categories="news")
news_vocab=set(word.lower() for word in news_words)
print("total words",len(news_words))
print('vocabulary size',len(news_vocab))
print('sample vocab',list(news_vocab)[:30])
```
