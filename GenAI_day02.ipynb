```python
import nltk
import json
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# -------------------------------
# Function to create corpus
# -------------------------------
def create_corpus_from_text(text):
    text = text.lower()
    tokens = word_tokenize(text)
    return tokens

# -------------------------------
# Function to clean corpus
# -------------------------------
def clean_corpus(tokens):
    stop_word = set(stopwords.words("english"))
    clean_tokens = [
        word for word in tokens
        if word not in stop_word and word not in string.punctuation
    ]
    return clean_tokens

# -------------------------------
# Function to generate vocabulary
# -------------------------------
def generate_voca(tokens):
    return sorted(set(tokens))

# -------------------------------
# Save vocabulary to JSON
# -------------------------------
def save_voca_to_json(voca, filename):
    with open(filename, "w") as f:
        json.dump(voca, f, indent=4)

# -------------------------------
# Load vocabulary from JSON
# -------------------------------
def load_voca_from_json(filename):
    with open(filename, "r") as f:
        return json.load(f)

# -------------------------------
# Main Execution
# -------------------------------
if __name__ == "__main__":
    user_text = """
    Natural language processing is a subfield of AI that focuses on the
    interaction between computers and human language.
    """

    corpus_tokens = create_corpus_from_text(user_text)
    clean_tokens = clean_corpus(corpus_tokens)
    voca = generate_voca(clean_tokens)

    save_voca_to_json(voca, "voca.json")
    loaded_voca = load_voca_from_json("voca.json")

    print("\nClean Tokens:")
    print(clean_tokens)

    print("\nVocabulary Loaded from JSON:")
    print(loaded_voca)

```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import (
    sent_tokenize,
    word_tokenize,
    WhitespaceTokenizer,
    RegexpTokenizer,
    TreebankWordTokenizer,
    TweetTokenizer
)

text = "Hello Umang! This is an NLP tutorial. Let's learn tokenization using NLTK."
tweet_text = "lovng NLP!!! 😎😉 check https://nltk.org @nltk_org #NLProc"


print(text)

print(sent_tokenize(text))

wt = WhitespaceTokenizer()
print(wt.tokenize(text))


rt = RegexpTokenizer(r'\w+')
print(rt.tokenize(text))

treebank_tok = TreebankWordTokenizer()
print(treebank_tok.tokenize("they'll save"))

print(tweet_text)

tt = TweetTokenizer()
print(tt.tokenize(tweet_text))
```
